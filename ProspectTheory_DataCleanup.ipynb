```python
import pandas as pd
import numpy as np
```

```python
csv_file_one = "FAZA1_SIMULACIJA_BACKUP.csv"
csv_file_two = "FAZA2_SIMULACIJA_BACKUP.csv"
# df = pd.read_csv(csv_file, sep=',', encoding='utf8')
```

```python
df_one = pd.read_csv(csv_file_one, sep=',', encoding='utf8', index_col='_id')
df_two = pd.read_csv(csv_file_two, sep=',', encoding='utf8', index_col='_id')
```

```python
df = df_one
```

```python
df_answers = df.filter(regex='answers')
```

```python
df.drop(list(df.filter(regex = 'answers')), axis = 1, inplace = True)
df.head()
```

```python
df_answers.head()
```

```python
df_answers_new = df_answers
```

```python
for col in df_answers.columns:
    taskName = col.split('.')[1]
    
    answer = df_answers[col]
    
    df_answers_new[col] = answer.str.replace("\[|\]|\'", "", regex=True).str.split(",")
    
    split_df = pd.DataFrame(df_answers_new[col].values.tolist(), columns = ['answer.{}'.format(taskName), 'time.{}'.format(taskName)])
    
    df = df.concat(split_df)
```
